## Summary statistics: mean, median, mode, variance, standard deviation, standard error

```python

def mean(x: List) -> float:
	if not isinstance(x, list):
		raise ValueError("Input must be a list")
	if x == []:
		return
	return sum(x)/len(x)

def median(x: List) -> float:
	if not isinstance(x, list):
		raise ValueError("Input must be a list")
	if x == []:
		return
	x_copy = x.copy().sort()
	x_copy.sort()
	return x_copy[len(x_copy)//2]

def mode(x):
	return Counter(x).most_common(1)[0][0]

def variance(x: List) -> float:
	if not isinstance(x, list):
		raise ValueError("Input must be a list")
	if x == []:
		return
	x_mean = mean(x)
	return sum((v - x_mean)**2 for v in x)/(len(x)-1)

def standard_deviation(x: List) -> float:
	return variance(x)**0.5
```

## Ranking

```python
# Competitive Ranking (1224)
def competitive_ranking(scores: List[int]) -> List[int]:
    if not isinstance(scores, list):
        raise ValueError("Input must be a list")
    if scores == []:
        return []
    score_to_rank = {}
    for i, score in enumerate(sorted(set(scores), reverse=True)):
        score_to_rank[score] = i + 1
    return [score_to_rank[score] for score in scores]

# Dense Ranking (1223)
def dense_ranking(scores: List[int]) -> List[int]:
    if not isinstance(scores, list):
        raise ValueError("Input must be a list")
    if scores == []:
        return []
    score_to_rank = {}
    rank = 1
    for score in sorted(set(scores), reverse=True):
        score_to_rank[score] = rank
        rank += 1
    return [score_to_rank[score] for score in scores]

# Ordinal Ranking (1234)
def ordinal_ranking(scores: List[int]) -> List[int]:
    if not isinstance(scores, list):
        raise ValueError("Input must be a list")
    if scores == []:
        return []
    score_to_rank = {}
    for i, score in enumerate(sorted(set(scores), reverse=True)):
        score_to_rank[score] = i + 1
    return [score_to_rank[score] for score in scores]

# Fractional Ranking (1.5, 1.5, 3)
def fractional_ranking(scores: List[int]) -> List[float]:
    from collections import Counter

    if not isinstance(scores, list):
        raise ValueError("Input must be a list")
    if scores == []:
        return []
    score_to_rank = {}
    for i, score in enumerate(sorted(set(scores), reverse=True)):
        score_to_rank[score] = i + 1
    rank_counts = Counter(scores)
    fractional_ranks = {}
    for score, count in rank_counts.items():
        start_rank = score_to_rank[score]
        end_rank = start_rank + count - 1
        fractional_ranks[score] = (start_rank + end_rank) / 2
    return [fractional_ranks[score] for score in scores]

def rank(scores: List[int], method: str) -> List[float]:
    if method == "competitive":
        return competitive_ranking(scores)
    elif method == "dense":
        return dense_ranking(scores)
    elif method == "ordinal":
        return ordinal_ranking(scores)
    elif method == "fractional":
        return fractional_ranking(scores)
    else:
        raise ValueError("Invalid ranking method")
```

## Measures of association: covariance, correlation

```python
def cov(x: List, y: List) -> float:
	if not isinstance(x, list) or not isinstance(y, list):
		raise ValueError("Inputs must be lists")
	x_centered = [v - mean(x) for v in x]
	y_centered = [v - mean(y) for v in y]
	return sum(i*j for i,j in zip(x_centered, y_centered))/(len(x)-1)

def corr(x,y):
	x_stdev = standard_deviation(x)
	y_stdev = standard_deviation(y)
	return cov(x, y)/(x_stdev * y_stdev)

def spearman_corr(x, y):
	x_ranked = rank(x, method="fractional")
	y_ranked = rank(y, method="fractional")
	return corr(x_ranked, y_ranked)

n = 1000
x = [random.randint(1, 10) for _ in range(n)]
y = [random.randint(1, 10) for _ in range(n)]

x_ranked = rank(x, method="fractional")
y_ranked = rank(y, method="fractional")

print(abs(spearman_corr(x,y) - spearmanr(x,y).statistic))
assert abs(spearman_corr(x,y) - spearmanr(x,y).statistic) < 1e-6
```

## Matrix operations: addition, multiplication, transpose, inverse

```python
def matmul(X, Y):
    ix = len(X)
    jx = len(X[0])
    iy = len(Y)
    jy = len(Y[0])
    nrows, ncols = ix, jy
    if jx != iy:
        raise ValueError("Incompatible dimensions for matrix multiplication")
    res = [[0] * ncols for _ in range(nrows)]
    for i in range(nrows):
        for j in range(ncols):
            for k in range(jx):
                res[i][j] += X[i][k] * Y[k][j]
    return res

A = [[1, 2], [3, 4]]
B = [[1, 0], [0, 1]]
B = [[0, 1], [1, 0]]
B = [[1, 0], [0, -1]]
print(A)
print(B)
print(matmul(A, B))
```

## Fibonacci sequence

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)
[fib(i) for i in range(10)]
```

```python
def fib_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fib_memo(n-1, memo) + fib_memo(n-2, memo)
    return memo[n]
[fib_memo(i) for i in range(10)]
```

## API Simulation

```python
import time
import random
import json

def api_get(prompt):
    # Simulate API call with random success/failure
    if random.random() < 0.8:  # 80% chance of success
        return {"response": f"LLM response to: {prompt}"}
    else:
        raise Exception("API rate limit exceeded")

def robust_call_llm_api(prompt, max_retries=5):
    retries = 0
    while retries < max_retries:
        try:
            response = api_get(prompt)
            return response
        except Exception as e:
            print(f"Error: {e}. Retrying in {2 ** retries} seconds...")
            time.sleep(2 ** retries)  # Exponential backoff
            retries += 1
    raise Exception("Max retries exceeded")

# Example usage:
prompt = "What is the capital of France? Answer in JSON format: {'response': <answer>}"
response = robust_call_llm_api(prompt)
print(json.dumps(response, indent=2))
```

## Calculate the $n$th quantile of a list of numbers.

```python
def quantile(X: List, q: float) -> float:
    """
    Calculate the q-th quantile of a list of numbers.
    Works by sorting the list and using linear interpolation between the two closest ranks.
    """
    if not isinstance(X, list):
        raise ValueError("Input must be a list")
    if X == []:
        return None
    if not (0 <= q <= 1):
        raise ValueError("Quantile must be between 0 and 1")
    X_sorted = sorted(X)
    n = len(X_sorted)
    index = q * (n - 1) # 0-based index
    lower = int(index) # floor of index
    upper = min(lower + 1, n - 1) # ceiling of index, capped at n-1
    weight = index - lower # fractional part of index
    # Linear interpolation between the two closest ranks
    return X_sorted[lower] * (1 - weight) + X_sorted[upper] * weight 

quantile([1, 2, 3, 4, 5], 0.25)
```

## Bootstrap p-values, confidence intervals, and standard errors

### difference of means between two samples

```python
x = [13, 12, 7, 16, 9, 11, 7, 10, 9, 8, 9, 7, 16, 7, 9,
8, 13, 10, 11, 9, 13, 13, 10, 10, 9, 7, 7, 6, 7, 8, 12, 13, 9, 6, 9, 11, 10, 8, 12, 10, 9, 10, 8, 14, 13, 13, 10, 11]

y = [8, 8, 16, 16, 9, 13, 14, 13, 10, 12, 10, 6, 14, 8,
13, 14, 7, 13, 7, 8, 4, 11, 7, 12, 8, 9, 12, 8, 11, 10, 12, 6, 10, 15, 11, 12, 3, 8, 11, 10, 10, 8, 12]

pop = x + y

n_boots = 50_000
boot_diffs = []
for _ in range(n_boots):
    boot_x = random.choices(population=pop, k=len(x)) 
    boot_y = random.choices(population=pop, k=len(y)) 
    boot_diffs.append(mean(boot_x) - mean(boot_y))
print("CI:", quantile(boot_diffs, 0.025), quantile(boot_diffs, 0.975))
print("CI:", np.quantile(boot_diffs, 0.025), np.quantile(boot_diffs, 0.975))

# p-value
obs_diff = mean(x) - mean(y)
p_value = sum(1 for d in boot_diffs if abs(d) >= abs(obs_diff)) / n_boots
print("p-value:", p_value)
```

### Variance between two samples

```python
import numpy as np
import random

x = [13, 12, 7, 16, 9, 11, 7, 10, 9, 8, 9, 7, 16, 7, 9,
8, 13, 10, 11, 9, 13, 13, 10, 10, 9, 7, 7, 6, 7, 8, 12, 13, 9, 6, 9, 11, 10, 8, 12, 10, 9, 10, 8, 14, 13, 13, 10, 11]

y = [8, 8, 16, 16, 9, 13, 14, 13, 10, 12, 10, 6, 14, 8,
13, 14, 7, 13, 7, 8, 4, 11, 7, 12, 8, 9, 12, 8, 11, 10, 12, 6, 10, 15, 11, 12, 3, 8, 11, 10, 10, 8, 12]

pop = x + y

n_boots = 50_000

diff = np.var(y, ddof=1) - np.var(x, ddof=1)
print('observed difference', diff)

counts = 0
pop_boot = pop.copy()
for _ in range(n_boots):

    x_boot = random.choices(population=pop, k=len(x))
    y_boot = random.choices(population=pop, k=len(y))

    diff_boot = np.var(y_boot, ddof=1) - np.var(x_boot, ddof=1)

    if diff_boot >= diff:
        counts += 1

p_value = counts/n_boots
print(counts, p_value)
```

## k-NN

```python
k = 3
n = 20
x = [(random.random(), random.random()) for _ in range(n)]
y = [3.1*i + 2.7*j for i,j in x]
x0 = (random.random(), random.random())
x0
plt.figure()
plt.scatter([v[0] for v in x], [v[1] for v in x])
plt.scatter(x0[0], x0[1], color='red')
plt.show()
```

```python
def kNN(x, y, x0, k):
    n = len(x)
    distances = [((v[0] - x0[0])**2 + (v[1] - x0[1])**2)**0.5 for v in x]
    nearest_indices = sorted(range(n), key=lambda i: distances[i])[:k] # A trick to remember!
    return sum(y[i] for i in nearest_indices)/k

kNN(x, y, x0, k)
```

```python
# compute distances between x0 and all x' in x
def dist(x0, x):
    d = []
    for v in x:
        d.append(( (x0[0] - v[0])**2 + (x0[1] - v[1])**2)**0.5)
    return d

def min_dist(x0, x):
    distances = dist(x0, x)
    m = min(distances)
    return distances.index(m)

def argmin_dist(x0, x):
    m = min_dist(x0,x)
    return x[m]

argmin_dist(x0,x)

def knn(x, y, x0, k):
    neighbors = []
    x_copy = x.copy()
    for _ in range(k):

        neighbor = argmin_dist(x0, x_copy)
        neighbors.append(neighbor)
        x_copy.remove(neighbor)
    return y

knn(x=x, x0=x0, k=3)
```
