## Bootstrap and Poisson Bootstrap

See https://www.unofficialgoogledatascience.com/2015/08/an-introduction-to-poisson-bootstrap26.html

```python
import random

B = 3
X = [2.3, 1.1, 2.7, 4]
n = len(X)
X_bar = sum(X)/n
print("X_bar: ", X_bar)
# Simulate X* of size n, by sampling from X with replacement
bootstrap_samples = []
for j in range(B):
	X_boot = []
	for i in range(n):
		idx = random.randint(0, n-1)
		X_boot.append(X[idx])
	bootstrap_samples.append(X_boot)

print(X_boot)
print("Bootstrapped samples:", bootstrap_samples)
bootstrap_means = []
for X_boot in bootstrap_samples:
	bootstrap_means.append(sum(X_boot)/n)

print(bootstrap_means)
```

```python

X_bar_boot = sum(bootstrap_means)/B
X_se_boot = (sum([(x-X_bar)**2 for x in bootstrap_means])/(B-1))**0.5
print(X_bar, X_bar_boot, X_se_boot)
print(X_bar - 1.96*X_se_boot, X_bar + 1.96*X_se_boot)
```

```python
# Poisson Bootstrap: Convert to counts
# [c(X[0]), ..., c(X[n-1])]
X = [2.3, 1.1, 2.7, 4]

# convert list of floats to freq table
from typing import List
from collections import defaultdict

def list_to_frequencies(X_boot: List, X: List) -> list:
    X_boot_frequencies = defaultdict(int)
    for x_i in X_boot:
        X_boot_frequencies[x_i] += 1

    # put in lists in canonical order
    X_boot_frequencies_canonical = []
    for x_i in X:
        if x_i in X_boot_frequencies:
            X_boot_frequencies_canonical.append(X_boot_frequencies[x_i])
        else:
            X_boot_frequencies_canonical.append(0)
    return X_boot_frequencies_canonical

for X_boot in bootstrap_samples:
    X_boot_frequencies_canonical = list_to_frequencies(X_boot, X)
    print(X_boot, '-->', X_boot_frequencies_canonical)
```

```python
# Convert to Poisson counts
frequencies = []
for X_boot in bootstrap_samples:
    X_boot_frequencies_canonical = list_to_frequencies(X_boot, X)
    print(X_boot, '-->', X_boot_frequencies_canonical)
    frequencies.append(X_boot_frequencies_canonical)

frequencies

poisson_counts = defaultdict(int)
for freqs_list in frequencies:
    for i, fi in enumerate(freqs_list):
        xi = X[i]
        poisson_counts[xi] += 1
poisson_counts
```
